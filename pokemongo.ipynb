```python
import pokemongo as pg

my_pokemon = pg.read_csv("""
Pokemon      CP  Move1        Move2
Arcanine   1522  FireFang     Bulldoze
Arcanine   1356  FireFang     Flamethrower
Rhydon     1192  MudSlap      Earthquake
Pinsir     1118  FuryCutter   Submission
Pidgeot    1094  WingAttack   Hurricane
Rhydon     1057  MudSlap      Megahorn
Sandslash  1031  MudShot      Earthquake
Magmar     1026  Ember        Flamethrower
Arbok       981  Acid         SludgeWave
Weepinbell  950  Acid         PowerWhip
Golbat      943  Bite         OminousWind
Flareon     927  Ember        HeatWave
Tauros      919  ZenHeadbutt  Earthquake
Ivysaur     873  VineWhip     SludgeBomb
Fearow      869  Peck         DrillRun
Lickitung   854  Lick         Stomp
Venomoth    840  Confusion    PoisonFang
Graveler    804  MudShot      RockSlide
Marowak     779  MudSlap      Dig
Raticate    768  QuickAttack  HyperFang
Jolteon     766  ThunderShock Thunder
Primeape    751  KarateChop   LowSweep
Aerodactyl  677  Bite         IronHead
Growlithe   592  Ember        BodySlam
Machoke     584  LowKick      BrickBreak
Dodrio      562  FeintAttack  AerialAce
Squirtle    561  Bubble       AquaTail
""")

my_pokemon['DPS1'] = my_pokemon.apply(lambda p: pg.get_dps(p, p.Move1), axis=1)
my_pokemon['DPS2'] = my_pokemon.apply(lambda p: pg.get_dps(p, p.Move2), axis=1)
my_pokemon.sort_values('DPS1', ascending=False)
```
